```python
a = open('lng.txt', 'w')
lines =[]
for i in data:
    lng = geocoder.google(i).lng
    lat = geocoder.google(i).lat
    print(i, lng, lat)
    if str(lng) != 'None':
        lines.append((str(lat), str(lng)))
    else:
        lines.append(i)

a.writelines(lines)
    
```
